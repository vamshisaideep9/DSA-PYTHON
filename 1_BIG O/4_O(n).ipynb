````python
import time
````

````python
nemo = ['nemo']
everyone = ['dory', 'bruce', 'marlin', 'nemo', 'gill', 'bloat', 'nigel', 'squirt', 'darla']
large = ['nemo' for i in range(100000)]
````

````python
def find_nemo(array):
    t0 = time.time() #o(1)
    for i in range(0,len(array)): #O(m)
        if array[i] == 'nemo': #m*O(1)
            print("Found Nemo") #k1*O(1)
    t1 = time.time() #O(1)
    print(f'The search took {t1-t0} seconds')
````

````python
find_nemo(nemo)
find_nemo(everyone)
find_nemo(large)
````

````python
def funchallenge(input):
    temp = 10 #O(1)
    temp = temp + 50 #O(1)
    for i in range(len(input)): #O(n)
        var = True #n*O(1)
        temp += 1 #n*O(1)
    return temp #O(1)

````

````python
funchallenge(nemo)
````

````python
funchallenge(everyone)
````

````python
funchallenge(large)
````

```python
Total running time of the funchallenge function = 

O(1+1+n+n*1+n*1+n*1+1) = O(3n+n) = O(3(n+1))

Any constant in Big-O representation can be replaced by 1, as it dosent really matter what constant it is.

Therefore, O(3(n+1)) becomes O(n+1)

Similarly, any constant number added or subtracted to n or multiplied or divided by n can also be safely written as just n

This is because the constant that operates upon n, doesnt depend on n, i.e., the input size

Therefore, the funchallenge function can be said to be of O(n) or Linear Time Complexity.
```
